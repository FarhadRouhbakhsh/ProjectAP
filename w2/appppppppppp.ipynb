```python
import requests
import pandas as pd
import numpy as np
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
r = requests.get(url)
json = r.json()
json.keys()
```

```python
elements_df = pd.DataFrame(json['elements'])
teams_df = pd.DataFrame(json['teams'])
```

```python
elements_df.keys()
```

```python
posts = elements_df.groupby('element_type')
print(posts.groups)
```

```python
GK = posts.get_group(1)
DF = posts.get_group(2)
MD = posts.get_group(3)
ST = posts.get_group(4)
GK
```

```python
pd.options.mode.chained_assignment = None
GK['defence']=0
for row in GK.index:
    team_id=GK['team'][row]
    team_row=GK.loc[GK['id']==team_id]
    strength=team_row['strength_defence_home']
    GK['defence'][row]=strength
GK
```
