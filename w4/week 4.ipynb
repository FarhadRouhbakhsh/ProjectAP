```python
import requests
import pandas as pd
import numpy as np
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
r = requests.get(url)
json = r.json()
json.keys()
```

```python
elements_df = pd.DataFrame(json['elements'])
teams_df = pd.DataFrame(json['teams'])
```

```python
standard=pd.read_excel("C://Users//farha//Downloads//sportsref_download (1).xlsx",header=1)
```

```python
standard=standard.sort_values('Player')# remove players from 2 team
standard.index=list(range(len(standard.index)))
for row in standard.index:
    if row>0:
        try:
            if standard['Player'][row]==standard['Player'][row-1]:
                standard=standard.drop(index=row)
        except:
            continue        
standard.index=list(range(len(standard.index)))
standard.loc[standard['Player']=='Ross Barkley']
```

```python
for row in teams_df.index: #changing team names in fantasy stats to use in fbref
    team_name=teams_df['name'][row]
    if team_name=='Leicester':
        teams_df['name'][row]='Leicester City'
    if team_name=='Man City':
        teams_df['name'][row]='Manchester City'
    if team_name=='Man Utd':
        teams_df['name'][row]='Manchester Utd'
    if team_name=='Newcastle':
        teams_df['name'][row]='Newcastle Utd'
    if team_name=='Spurs':
        teams_df['name'][row]='Tottenham'
    if team_name=='Leeds':
        teams_df['name'][row]='Leeds United'
teams_df
```

```python
elements_df['team_name']='' #adding team name to elements_df
for row in elements_df.index: 
    team_id=elements_df['team'][row]
    team_row=teams_df.loc[teams_df['id']==team_id]
    team_row.index=[0]
    team_name=team_row['name'][0]
    elements_df['team_name'][row]=team_name
```

```python
players=elements_df
for row in elements_df.index: # remove players who haven't played yet.
    if elements_df['minutes'][row]==0.:
        players=players.drop(index=row)
players.index=list(range(len(players.index)))
players
```

```python
for row in players.index: #Updating position slot
    pos=players['element_type'][row]
    if pos==1:
        players['element_type'][row]='GK'
    if pos==2:
        players['element_type'][row]='DF'
    if pos==3:
        players['element_type'][row]='MF'
    if pos==4:
        players['element_type'][row]='FW'
players[['web_name','element_type']]
```

```python
#correct bad names
standard.loc[standard['Player']=='Jóhann Berg Guðmundsson','Player']='Johann Berg Gudmundsson'
standard.loc[standard['Player']=='İlkay Gündoğan','Player']='Ilkay Gündogan'
standard.loc[standard['Player']=='Carlos','Player']='Carlos Vinicius'
standard.loc[standard['Player']=='Łukasz Fabiański','Player']='Lukasz Fabianski'
standard.loc[standard['Player']=='Tomáš Souček','Player']='Tomas Soucek'
standard.loc[standard['Player']=='Dan Nlundulu','Player']="Daniel N'Lundulu"
players['name']=''
for row in players.index:
    checked=False 
    team_name=players['team_name'][row]
    first_name=players['first_name'][row]
    second_name=players['second_name'][row]
    web_name=players['web_name'][row]
    pos=players['element_type'][row]
    if first_name in web_name: #build a name for checking
        name=web_name
    else:
        name=first_name+" "+web_name
    players['name'][row]=name
    for s_row in standard.index:
        fb_name=standard['Player'][s_row]
        fb_team=standard['Squad'][s_row]
        fb_pos=standard['Pos'][s_row]
        if name==fb_name or first_name+' '+second_name==fb_name: #find names which are totally equal
            checked=True
            #print(name,' = ',fb_name,end='\n')
    if checked==False: #find names which are not totally equal
        for s_row in standard.index: 
            fb_name=standard['Player'][s_row]
            fb_team=standard['Squad'][s_row]
            fb_pos=standard['Pos'][s_row]
            if team_name==fb_team:
                if web_name in fb_name or second_name in fb_name:
                    #print(name,' = ',fb_name,end='\n')
                    standard.loc[standard['Player']==fb_name,'Player']=name
                    checked=True
                elif first_name in fb_name and pos in fb_pos:
                    standard.loc[standard['Player']==fb_name,'Player']=name
                    checked=True
                    #print(name,' = ',fb_name,end='\n')
players
```

```python
posts = players.groupby('element_type')
posts.groups
```

```python
GK = posts.get_group('GK')
DF = posts.get_group('DF')
MF = posts.get_group('MF')
FW = posts.get_group('FW')
```

```python
pd.options.mode.chained_assignment = None
def add_cl(POST,DF,CL):
    POST[CL] = 0 #add new coloumn to the end
    for row in POST.index:
        team_id = POST['team'][row] #get team id from post
        team_row = DF.loc[DF['id'] == team_id] #get specific row for team id 
        POST[CL][row] = team_row[CL] #update coloumn
    return POST
```

```python
add_cl(GK,teams_df,'strength_defence_home')
add_cl(GK,teams_df,'strength_defence_away')
add_cl(DF,teams_df,'strength_attack_home')
add_cl(DF,teams_df,'strength_attack_away')
add_cl(DF,teams_df,'strength_defence_home')
add_cl(DF,teams_df,'strength_defence_away')
add_cl(MF,teams_df,'strength_attack_home')
add_cl(MF,teams_df,'strength_attack_away')
add_cl(MF,teams_df,'strength_defence_home')
add_cl(MF,teams_df,'strength_defence_away')
add_cl(FW,teams_df,'strength_attack_home')
add_cl(FW,teams_df,'strength_attack_away')
```
